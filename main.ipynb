# Preprocessing

The dataset is composed of about 8000 images taken with different consumer cameras. For this reason, the images are affected by high variability in illumination, angle, quality, size, blurriness, etc. Furthermore, the regions which contain the skin disease are not isolated.

The algorithm used to process such images is adapted by from the one developed by Alessandro Dâ€™Amico, Riccardo Murgia and Mazeyar Moeini Feizabadi [https://github.com/eskinderit/experiments-synthetic-generation-clinical-skin-images/blob/main/generate_images.ipynb . The functions used in the next cells are imported from the 'generate_images.py' file.

The idea is to detect the regions of the image in which the skin disease is present by considering the binary mask of the image and employing a sliding-window approach. In particular:

The algorithm starts from the top-left corner of the image and takes a fixed-size patch (in our case the size is 256x256).

The algorithm extracts the binary mask relative to that patch and computes the disease coverage in that patch, i.e. the ratio of the positive label (1) in the mask over the negative label (0). This is done to assess the presence of contrast, which could indicate that the skin disease is present. If the disease coverage is higher than a threshold (high contrast), the patch is taken. If it is lower, the patch is discarded.

Then, it repeats the process by sliding the patch extractor by a certain amount. This process results in a lot of patches, the majority of which overlap and, thus, are redundant. To overcome this issue, a non-maxima suppression procedure is run over the obtained patches: if two patches overlap over a certain limit, the patch with lower disease coverage is discarded. The last step involves discarding those crops which exhibit low contrast, in hope of cleaning the dataset from poorly illuminated or blurry patches.

# Skin Tone distribution across the three diseases

Here, we visualize the distribution of the six different skin tones across the three diseases, in order to have an idea of how many synthetic images of darker skin tones should be generated.

```python
folder_path = "data"
```

```python
dataset = CustomImageFolderColor(folder_path)
print("Len dataset: ", len(dataset), flush=True)

# Split the dataset into train and test sets
train_size = int(0.6 * len(dataset))  # 60% train, 20% validation, 20% test
val_size = int(0.2 * len(dataset))
test_size = len(dataset) - train_size - val_size
train_dataset, val_dataset, test_dataset = random_split(dataset, [train_size, val_size, test_size])
```

```python
def analyze_dataset(train_dataset):
    labels = [0, 1, 2]
    skin_labels = ['dark', 'brown', 'tan', 'intermediate', 'light', 'very light']
    
    # Dictionary to store counts for each label and skin_label combination
    label_skin_counts = {label: Counter() for label in labels}
    total_counts = Counter()
    
    # Iterate over the dataset
    for image, skin_label, label in tqdm(train_dataset):
        if label in labels and skin_label in skin_labels:
            label_skin_counts[label][skin_label] += 1
            total_counts[label] += 1
    
    # Print results and plot histograms
    for label in labels:
        print(f"Results for label {label}:")
        total = total_counts[label]
        counts = []
        for skin_label in skin_labels:
            count = label_skin_counts[label][skin_label]
            percentage = (count / total * 100) if total > 0 else 0
            counts.append(count)
            print(f"  {skin_label}: {count} ({percentage:.2f}%)")
        print("-")
        
        # Plot histogram
        plt.figure(figsize=(8, 5))
        bars = plt.bar(skin_labels, counts, color='skyblue')
        plt.xlabel("Skin Label")
        plt.ylabel("Count")
        plt.title(f"Distribution of Skin Labels for disease {diseases[label]}")
        plt.xticks(rotation=45)
        
        # Add count labels on top of bars
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{height}', ha='center', va='bottom')
        
        plt.show()

analyze_dataset(train_dataset)
```

```python
LEARNING_RATE = 0.01
BATCH_SIZE = 128
EPOCHS = 30
device = 'cuda'
```

```python
dataset = CustomImageFolderColor(folder_path)
print("Len dataset: ", len(dataset), flush=True)

# Split the dataset into train and test sets
train_size = int(0.6 * len(dataset))  # 60% train, 20% validation, 20% test
val_size = int(0.2 * len(dataset))
test_size = len(dataset) - train_size - val_size
train_dataset, val_dataset, test_dataset = random_split(dataset, [train_size, val_size, test_size])
```

```python
# Create data loaders for train, validation and test sets
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
validation_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle= False)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

############################### INSTANTIATING THE MODEL #################################################
net = Net(3)

if torch.cuda.device_count() > 1:
  print("Let's use", torch.cuda.device_count(), "GPUs!", flush=True)
  # dim = 0 [30, xxx] -> [10, ...], [10, ...], [10, ...] on 3 GPUs
  model = nn.DataParallel(net)

net = net.to(device)
print(net, flush=True)

############################### DEFINING THE LOSS, THE OPTIMIZER AND THE SCHEDULER #####################################
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(net.parameters(), lr=LEARNING_RATE, momentum=0.9)
scheduler = lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS)
early_stopper = EarlyStopper(patience=3, min_delta=0.03)

############################### TRAINING THE MODEL ######################################################
train(net, FLAG, EPOCHS, LEARNING_RATE, BATCH_SIZE, optimizer, criterion, scheduler, early_stopper, train_loader, validation_loader)
```

## Evaluation

### Accuracy and F1 scores

```python
net.eval()

classes = ['0', '1', '2']
skin_colors = ['dark', 'brown', 'tan', 'intermediate', 'light', 'very light']

# prepare to count predictions for each class for the computation of ACCURACY
correct_pred = {classname: {skin_color: 0 for skin_color in skin_colors} for classname in classes}
total_pred = {classname: {skin_color: 0 for skin_color in skin_colors} for classname in classes}

# prepare to store the predictions for each class for the computation of F1 SCORE
pred_F1 = {classname: {skin_color: [] for skin_color in skin_colors} for classname in classes}
labels_F1 = {classname: {skin_color: [] for skin_color in skin_colors} for classname in classes}

# prepare to store the predictions for the computation of the FAIRNESS METRICS
pred_fairness = []
labels_fairness = []
skin_labels_fairness = []

# again no gradients needed
with torch.no_grad():
    for data in tqdm(test_loader):
        images, skin_labels, labels = data
        images, labels = images.to(device), labels.to(device)
        outputs = net(images)
        _, predictions = torch.max(outputs, 1)
        # collect the correct predictions for each class and skin color for the computation of ACCURACY
        for label, skin_label, prediction in zip(labels.tolist(), skin_labels, predictions.tolist()):
            # store predictions and labels for each class and skin color for the computation oF F1 SCORE
            pred_F1[classes[label]][skin_label].append(prediction)
            labels_F1[classes[label]][skin_label].append(label)

            if label == prediction:
                correct_pred[classes[label]][skin_label] += 1
            total_pred[classes[label]][skin_label] += 1

            pred_fairness.append(prediction)
            labels_fairness.append(label)
            skin_labels_fairness.append(skin_label)
```

```python
print(f'Overall accuracy: {accuracy_score(labels_fairness, pred_fairness)*100:.1f}%', flush=True)
print(f"Overall F1 score: {f1_score(labels_fairness, pred_fairness, average='macro'):.2f}", flush=True)
```

```python
groups = np.array(['minority' if skin_labels_fairness[i] == 'dark' or skin_labels_fairness[i] == 'brown' else 'majority' for i in range(len(labels_fairness))])
pred_min = [pred_fairness[i] for i in range(len(pred_fairness)) if groups[i] == 'minority']
labels_min = [labels_fairness[i] for i in range(len(labels_fairness)) if groups[i] == 'minority']

pred_maj = [pred_fairness[i] for i in range(len(pred_fairness)) if groups[i] == 'majority']
labels_maj = [labels_fairness[i] for i in range(len(labels_fairness)) if groups[i] == 'majority']

print(f"Overall Minority F1: {f1_score(labels_min, pred_min, average='macro'):.2f},Overall Majority F1: {f1_score(labels_maj, pred_maj, average='macro'):.2f}", flush = True)
print(f"Overall Minority Accuracy: {accuracy_score(labels_min, pred_min)*100:.1f}%,Overall Majority Accuracy: {accuracy_score(labels_maj, pred_maj)*100:.1f}%", flush=True)
print("\n", flush=True)
```

```python
groups = ['minority' if skin_labels_fairness[i] == 'dark' or skin_labels_fairness[i] == 'brown' else 'majority' for i in range(len(labels_fairness))]

for disease_id in range(len(diseases)):
    correct_pred_min = 0
    correct_pred_maj = 0
    pred_min = 0
    pred_maj = 0

    for i in range(len(groups)):
        if labels_fairness[i] == disease_id:
            if groups[i] == 'majority':
                pred_maj += 1
                if labels_fairness[i] == pred_fairness[i]:
                    correct_pred_maj += 1
            elif groups[i] == 'minority':
                pred_min += 1
                if labels_fairness[i] == pred_fairness[i]:
                    correct_pred_min += 1

    print(f'Disease {diseases[disease_id]}: Minority Accuracy: {correct_pred_min/pred_min*100:.2f}%, Majority Accuracy: {correct_pred_maj/pred_maj*100:.2f}%', flush=True)

print("\n", flush=True)
```

```python
# Minority and Majority F1-score disease-wise

for disease_id in range(len(diseases)):
    pred_min = [1 if pred_fairness[i] == disease_id else 0 for i in range(len(pred_fairness)) if groups[i] == 'minority']
    labels_min = [1 if labels_fairness[i] == disease_id else 0 for i in range(len(labels_fairness)) if groups[i] == 'minority']

    pred_maj = [1 if pred_fairness[i] == disease_id else 0 for i in range(len(pred_fairness)) if groups[i] == 'majority']
    labels_maj = [1 if labels_fairness[i] == disease_id else 0 for i in range(len(labels_fairness)) if groups[i] == 'majority']

    print(f'Disease {diseases[disease_id]}: Minority F1: {f1_score(labels_min, pred_min):.2f}, Majority F1: {f1_score(labels_maj, pred_maj):.2f}', flush=True)

print("\n", flush=True)
```

### Fairness Metrics

**Disparate Impact**

```python
groups = ['minority' if skin_labels_fairness[i] == 'dark' or skin_labels_fairness[i] == 'brown' else 'majority' for i in range(len(labels_fairness))]

for disease_id in range(len(diseases)):
  # Initialize empty lists for minority and majority groups
  pred_list = []
  labels_list = []
  minority_list = []
  majority_list = []

  # Loop through skin_labels_fairness and check the group
  for i in range(len(pred_fairness)):
      pred_list.append(1 if pred_fairness[i] == disease_id else 0)
      labels_list.append(1 if labels_fairness[i] == disease_id else 0)

      if skin_labels_fairness[i] == 'dark' or skin_labels_fairness[i] == 'brown':
          minority_list.append(1)
          majority_list.append(0)
      else:
          minority_list.append(0)
          majority_list.append(1)

  print(f'Disparate impact for disease {diseases[disease_id]}: {disparate_impact(minority_list, majority_list, pred_list):.2f}', '\n', flush=True)
```

**Equalized Odds Ratio**

```python
groups = ['minority' if skin_labels_fairness[i] == 'dark' or skin_labels_fairness == 'brown' else 'majority' for i in range(len(labels_fairness))]

for disease_id in range(len(diseases)):
  preds = [1 if pred_fairness[i] == disease_id else 0 for i in range(len(pred_fairness))]
  labels = [1 if labels_fairness[i] == disease_id else 0 for i in range(len(labels_fairness))]

  print(f'Equalized Odds Ratio for {diseases[disease_id]}: {equalized_odds_ratio(labels, preds, sensitive_features=groups)}', '\n', flush=True)
```

## Evaluation

### Accuracy and F1 scores

```python
net.eval()

classes = ['0', '1', '2']
skin_colors = ['dark', 'brown', 'tan', 'intermediate', 'light', 'very light']

# prepare to count predictions for each class for the computation of ACCURACY
correct_pred = {classname: {skin_color: 0 for skin_color in skin_colors} for classname in classes}
total_pred = {classname: {skin_color: 0 for skin_color in skin_colors} for classname in classes}

# prepare to store the predictions for each class for the computation of F1 SCORE
pred_F1 = {classname: {skin_color: [] for skin_color in skin_colors} for classname in classes}
labels_F1 = {classname: {skin_color: [] for skin_color in skin_colors} for classname in classes}

# prepare to store the predictions for the computation of the FAIRNESS METRICS
pred_fairness = []
labels_fairness = []
skin_labels_fairness = []

# again no gradients needed
with torch.no_grad():
    for data in tqdm(test_loader):
        images, skin_labels, labels = data
        images, labels = images.to(device), labels.to(device)
        outputs = net(images)
        _, predictions = torch.max(outputs, 1)
        # collect the correct predictions for each class and skin color for the computation of ACCURACY
        for label, skin_label, prediction in zip(labels.tolist(), skin_labels, predictions.tolist()):
            # store predictions and labels for each class and skin color for the computation oF F1 SCORE
            pred_F1[classes[label]][skin_label].append(prediction)
            labels_F1[classes[label]][skin_label].append(label)

            if label == prediction:
                correct_pred[classes[label]][skin_label] += 1
            total_pred[classes[label]][skin_label] += 1

            pred_fairness.append(prediction)
            labels_fairness.append(label)
            skin_labels_fairness.append(skin_label)
```

```python
print(f'Overall accuracy: {accuracy_score(labels_fairness, pred_fairness)*100:.1f}%', flush=True)
print(f"Overall F1 score: {f1_score(labels_fairness, pred_fairness, average='macro'):.2f}", flush=True)
```
